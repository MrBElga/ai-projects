```python
preco_real = dados.iloc[:,0].values
```

```python
atributos = dados.iloc[:,1:12].values
```

```python
atributos[0:5,:]
```

```python
#0,1,3,5,8,9
from sklearn.preprocessing import LabelEncoder
labelencoder_atributos = LabelEncoder()
atributos[:,0] = labelencoder_atributos.fit_transform(atributos[:,0])
atributos[:,1] = labelencoder_atributos.fit_transform(atributos[:,1])
atributos[:,3] = labelencoder_atributos.fit_transform(atributos[:,3])
atributos[:,5] = labelencoder_atributos.fit_transform(atributos[:,5])
atributos[:,8] = labelencoder_atributos.fit_transform(atributos[:,8])
atributos[:,9] = labelencoder_atributos.fit_transform(atributos[:,9])
atributos[:,10] = labelencoder_atributos.fit_transform(atributos[:,10])
```

```python
atributos[0:5,:]
```

```python
#one hot enconding
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

onehotencoder = ColumnTransformer([('one_hot_encoder', OneHotEncoder(), [1,5,8,9])],
                                  remainder='passthrough')

atributos = onehotencoder.fit_transform(atributos).toarray()
```

```python
atributos.shape
```

```python
atributos[0:5,:]
```

```python
from keras.models import Sequential
from keras.layers import Dense

# Criação do modelo
regressor = Sequential()
regressor.add(Dense(units=25, activation='relu', input_dim=50))  # Ajuste o número de unidades e a dimensão de entrada para corresponder ao número de atributos
regressor.add(Dense(units=25, activation='relu'))
regressor.add(Dense(units=1, activation='linear'))
regressor.compile(loss='mean_absolute_error', optimizer='adam', metrics=[])
```

```python
regressor.fit(atributos, preco_real, batch_size=300, epochs=100)
```

```python
previsoes = regressor.predict(atributos)
```

```python
previsoes = previsoes.ravel()
```
