```python
import matplotlib.pyplot as plt
import numpy as np
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, BatchNormalization
from keras.utils import to_categorical
from sklearn.model_selection import StratifiedKFold
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
from sklearn.model_selection import StratifiedKFold
import numpy as np
```

```python
seed = 42
np.random.seed(seed)
```

```python
(X, y), (X_teste, y_teste) = mnist.load_data()
previsores = X.reshape(X.shape[0], 28, 28, 1).astype('float32') / 255.0
classe = to_categorical(y,10)
```

```python
kfold = StratifiedKFold(n_splits=4, shuffle=True, random_state=seed)
resultados = []
```

```python
for indice_treinamentos, indice_teste in kfold.split(previsores, np.zeros(shape=(classe.shape[0],1))):
  print('indice treino', indice_treinamentos, 'indice teste', indice_teste)
```

```python
classe.shape[0]
```

```python
for indice_treinamentos, indice_teste in kfold.split(previsores, np.zeros(shape=(classe.shape[0],1))):
    cnn = Sequential([
        Conv2D(32, (3, 2), input_shape=(28, 28, 1), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Conv2D(32, (3, 2), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
    ])

    cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    cnn.fit(previsores[indice_treinamentos], classe[indice_treinamentos], batch_size=450, epochs=20, validation_data=(previsores[indice_teste], classe[indice_teste]))


    previsoes = cnn.evaluate(previsores[indice_teste], classe[indice_teste])
    resultados.append(previsoes[1])
```
