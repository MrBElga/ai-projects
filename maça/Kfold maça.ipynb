```python
#carregar dados
import pandas as pd
dados = pd.read_csv('apple_quality.csv')
```

```python
  dados.columns
```

```python
from sklearn.preprocessing import LabelEncoder
atributos = dados.iloc[0:4000, 1:8].values
classe = dados.iloc[0:4000, 8].values

le = LabelEncoder()
classe_num = le.fit_transform(classe)
```

```python
#separar conjuntos
from sklearn.model_selection import train_test_split
treino, teste, classe_treino, classe_teste = train_test_split(atributos, classe_num, test_size=0.3, random_state=42)

```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout

# Função para criar o modelo
def criarRede():
    mlp = Sequential()
    mlp.add(Dense(units=16, activation='relu', kernel_initializer='random_uniform', input_dim=7))
    mlp.add(Dropout(0.2))
    mlp.add(Dense(units=16, activation='relu', kernel_initializer='random_uniform'))
    mlp.add(Dropout(0.2))
    mlp.add(Dense(units=1, activation='sigmoid'))
    mlp.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return mlp

```
