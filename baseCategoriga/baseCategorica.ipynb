```python
#Carregar dados
import pandas as pd
from keras.layers import Dense
from keras.utils import to_categorical
dados =  pd.read_csv('base_treinamento.csv')

```

```python
dados.columns
```

```python

from sklearn.preprocessing import LabelEncoder
atributos = dados.iloc[0:827, 0:6].values
classe = dados.iloc[0:827, 6:7].values

le = LabelEncoder()
classe_num = le.fit_transform(classe)
classe_one_hot = to_categorical(classe_num)
```

```python

from keras.models import Sequential
from keras.layers import Dense,Dropout
def criarRede():
    mlp = Sequential()
    mlp.add(Dense(units=6, activation='relu', kernel_initializer='random_uniform', input_dim=6))
    mlp.add(Dropout(0.2))
    mlp.add(Dense(units=6, activation='relu', kernel_initializer='random_uniform'))
    mlp.add(Dropout(0.2))
    mlp.add(Dense(units=5, activation='sigmoid'))

    mlp.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return mlp
```

```python
#aqui para salvar
from keras.callbacks import ModelCheckpoint
checkpoint = ModelCheckpoint(filepath='melhor.h5', monitor='categorical_acuraccy', save_best_only=True, mode='max', verbose=1)

```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score

## Criar o modelo KerasClassifier com o callback ModelCheckpoint
mlp = KerasClassifier(build_fn=criarRede, epochs=100, batch_size=10)

# Treinar o modelo com o callback ModelCheckpoint
resultado = cross_val_score(estimator=mlp, X=atributos, y=classe_one_hot, cv=10, scoring='accuracy', fit_params={'callbacks': [checkpoint]})
```
