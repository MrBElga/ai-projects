```python
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
```

```python
(X_treino, y_treino), (X_teste, y_teste) = mnist.load_data()
```

```python
plt.imshow(X_treino[10], cmap='gray')
```

```python
treino = X_treino.reshape(X_treino.shape[0], 28, 28, 1)
teste = X_teste.reshape(X_teste.shape[0], 28, 28, 1)

treino= treino.astype('float32')
teste= teste.astype('float32')

treino  /=255
teste /=255
```

```python
from keras.utils import to_categorical
classe_treino = to_categorical(y_treino,10)
classe_teste = to_categorical(y_teste,10)
```

```python
teste.shape[1:]
```

```python
cnn = Sequential()
cnn.add(Conv2D(32, (3,2), input_shape=(28,28,1),activation='relu'))
cnn.add(MaxPooling2D(pool_size=(2,2)))
cnn.add(Conv2D(32, (3,2), input_shape=(28,28,1),activation='relu'))
cnn.add(MaxPooling2D(pool_size=(2,2)))

cnn.add(Flatten())

cnn.add(Dense(units=405, activation='relu'))
cnn.add(Dense(units=405, activation='relu'))
cnn.add(Dense(units=10, activation='softmax'))

cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
cnn.summary()
```

```python
cnn.fit(treino, classe_treino, batch_size=405, epochs=20, validation_data=(teste,classe_teste))
```

```python
i = 9

plt.imshow(X_teste[i], cmap='gray')

saida = cnn.predict(X_teste[i].reshape(1,28,28,1))
print(saida)
import numpy as np

print('classe Resposta =',np.argmax(saida))
```
