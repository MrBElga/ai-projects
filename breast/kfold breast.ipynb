```python
#Carregar dados
import pandas as pd
dados = pd.read_csv('entradas_breast.csv')
classe = pd.read_csv('saidas_breast.csv')
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout

def criarRede():
  mlp = Sequential()
  mlp.add(Dense(units=16, activation='relu', kernel_initializer='random_uniform', input_dim=30))
  mlp.add(Dropout(0.2))
  mlp.add(Dense(units=16, activation='relu', kernel_initializer='random_uniform'))
  mlp.add(Dropout(0.2))
  mlp.add(Dense(units=1, activation='sigmoid'))

  mlp.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])
  return mlp
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score

mlp = KerasClassifier(build_fn=criarRede, epochs=100, batch_size=10)

resultado = cross_val_score(estimator=mlp, X=dados, y=classe, cv=10, scoring='accuracy')
```
